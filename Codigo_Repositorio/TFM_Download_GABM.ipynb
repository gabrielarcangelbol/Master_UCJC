#### 1.1.1.1 Descarga de muestra usando el endpoint Full-Archive API

```python
## Funcion para descargar tweets de acuerdo al limite mensual, haciendo uso de la API Premium de Twitter mediante Search Tweets API: Full Archive
'''
    The Full-archive endpoint on the premium Search Tweets API provides you complete and instant access to Tweets dating all the way back to the first Tweet in March 2006. 
    Tweets are available from the API approximately 30 seconds after being published.
    
    Documentacion completa: https://developer.twitter.com/en/docs/twitter-api/premium/search-api/api-reference/premium-search.
    Search Tweets: https://developer.twitter.com/en/docs/twitter-api/tweets/search/migrate/premium-to-twitter-api-v2.
    
    Parametros de la funcion:
    
    query: busqueda usando parametros como el lenguaje (lang), localidad (place, profile_country, profile_region, profile_locality), etiquetas (#), cuentas o menciones (@), enlaces (url)... entre otros, 
    fromDate: fecha inicial de busqueda (indicando hora de inicio),
    toDate: fecha final de busqueda (indicando hora de finalizacion),
    n: cantidad de publicaciones a retornar, 500 es el numero de maximo por llamada. Por ende es el que debe colocarse por si se quiere descargar mas data,
    next_token: None, si hay mas de 500 tweets por defecto el algoritmos buscara descargar hasta que no encuentre una paginacion siguiente o nos quedemos sin recuros.
'''
def search_tweets_fullarchive(query, fromDate, toDate, n, next_token):
    
    if next_token == None:
    
        header = {
            'authorization': headers['authorization'],
            'content-type': 'application/json'
        }
    
        data = '{"query":"'+query+ '","maxResults":"'+str(n)+'","fromDate":"'+fromDate+'","toDate":"'+ toDate+ '"}'
    

        url = 'https://api.twitter.com/1.1/tweets/search/fullarchive/FullArchiveUCJCPFMGABM.json'
    
        response = requests.post(url = url, headers=header, data=data).json()
    
        return(response)

    else:
        
        header = {
            'authorization': headers['authorization'],
            'content-type': 'application/json'
        }
    
        data = '{"query":"'+query+ '","maxResults":"'+str(n)+'","fromDate":"'+fromDate+'","toDate":"'+ toDate+ '","'+'next":"'+next_token+'"}'
    
        url = 'https://api.twitter.com/1.1/tweets/search/fullarchive/FullArchiveUCJCPFMGABM.json'
    
        response = requests.post(url = url, headers=header, data=data).json()
    
        return(response)                
```

#### 1.1.1.2 Descarga e insercion en la BD MongoDB usando el endpoint Full-Archive API

```python
## Funcion para descargar e insertar Tweets en BD mas de 30 dias de acuerdo al limite mensual, haciendo uso de la API Premium de Twitter mediante Search Tweets API: Full Archive 
'''
    The Full-archive endpoint on the premium Search Tweets API provides you complete and instant access to Tweets dating all the way back to the first Tweet in March 2006. 
    Tweets are available from the API approximately 30 seconds after being published.

    Documentacion completa: https://developer.twitter.com/en/docs/twitter-api/premium/search-api/api-reference/premium-search.
    Search Tweets: https://developer.twitter.com/en/docs/twitter-api/tweets/search/migrate/premium-to-twitter-api-v2.

    Parametros de la funcion:
    
    database: BD en la que se depositaran los datos en MongoDB, 
    query: busqueda usando parametros como el lenguaje (lang), localidad (place, profile_country, profile_region, profile_locality), etiquetas (#), cuentas o menciones (@), enlaces (url)... entre otros, 
    fromDate: fecha inicial de busqueda (sin indicar hora), 
    toDate: fecha final de busqueda (sin indicar hora), 
    clave: palabra o frase que se usa para identificar la coleccion que se inserta en MongoDB, 
    token: None, si hay mas de 500 tweets por defecto el algoritmos buscara descargar hasta que no encuentre una paginacion siguiente o nos quedemos sin recuros, 
    category: palabra clave que sirve para identificar la busqueda realizada.

'''
def search_tweets_fullarchive_completa(database, query, fromDate, toDate, clave , token, category):
    fromDate = fromDate.replace('-', '') + '0000'
    toDate = toDate.replace('-', '') + '2359'
    client = auth_mongo('XXXX', 'XXXX', database)    # XXXX - # Por motivos de seguridad se eliminaron estas direcciones / claves / puertos
    l = search_tweets_fullarchive(query = query, fromDate = fromDate, toDate = toDate,n = 100, next_token = token) # Api gratutita colocar en n=100, siendo 10 el valor minimom y 100 el maximo.
    if 'error' in str(l.keys()):
        regresiva(960)
    else:
        l["results"] = complete(l)
        df = pd.DataFrame(l['results'])
        df = Procesamiento_busqueda(df, clave)
        # insertando    
        df['word'] = query
        df['category'] = category
        insert_mongo(client, database, 't_'+clave+'_tag', df) 
        req = 1   
        print(req)
        regresiva(3)
        while "next" in l.keys():
            # El query
            print(l["next"])
            fic = open("token.txt", "w")
            fic.write(l["next"])
            fic.close()
            l = search_tweets_fullarchive(query = query, fromDate = fromDate, toDate = toDate,n = 100, next_token = l["next"]) # Api gratutita colocar en n=100, siendo 10 el valor minimom y 100 el maximo.
            if 'error' in str(l.keys()):
                regresiva(960)  
            else:
                l["results"] = complete(l)
                df = pd.DataFrame(l['results'])
                df = Procesamiento_busqueda(df, clave)
                df['word'] = query
                df['category'] = category
                # insertando los datos recopilados en la las llamadas en la BD definida en Mongo
                insert_mongo(client, database, 't_'+clave+'_tag', df)       
                req = req + 1 
                print(req)
                regresiva(3)
                if req == 2000:
                    break

                print("Loop Terminado")
```
